# Zero-shot Classification Pipeline

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "Advancements in AI are transforming healthcare.",
    candidate_labels=["health", "technology", "environment"],
)
```

# Text Generation Pipeline

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("Innovation is the driving force of our future. With each new breakthrough, we step closer to a world where")
```

# Custom Text Generation with Specific Model and Parameters

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "The secret to success in data science is",
    max_length=40,
    num_return_sequences=3,
)
```

# Fill-mask Pipeline

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("Artificial intelligence is transforming <mask> in unprecedented ways.", top_k=3)
```

# Named Entity Recognition (NER) Pipeline

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("Ada Lovelace and Alan Turing are pioneers in computer science, and they are both widely celebrated.")
```

#  Question-Answering Pipeline with Custom Context and Question

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

context = "Bumi adalah planet ketiga dari Matahari dan memiliki berbagai macam ekosistem."
question = "Planet ke berapakah Bumi dalam hubungan dengan Matahari?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

# Sentiment Analysis Pipeline

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("The new update made the app so much more user-friendly!")
```

# Summarization Pipeline

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    Machine learning is rapidly advancing with new techniques and models. Today,
    algorithms can recognize speech, diagnose diseases, and even predict human behavior.
    As these technologies evolve, they are beginning to be integrated into
    every facet of life, from healthcare to entertainment, marking a new era of
    technological innovation that will shape the future.
    """
)
```

# Translation Pipeline (Indonesian to English)

```python
from transformers import pipeline

translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Peningkatan teknologi sangat cepat akhir-akhir ini."
result = translator(text_to_translate)

print(result[0]['translation_text'])
```

# Analisis

## 1. Zero-shot Classification
Merupakan jenis klasifikasi dengan metode Zero-Shot yang menurut saya sangat menarik untuk dicoba karena tidak memerlukan data pelatihan di kategori tertentu. Metode ini cocok untuk analisis teks dalam skenario di mana kategori baru atau tidak terduga sering muncul, seperti mengkategorikan umpan balik pelanggan atau analisis tren media sosial. Zero-Shot memungkinkan kita mengklasifikasikan teks dengan cepat tanpa pelatihan khusus, yang mempercepat penerapannya di lingkungan produksi.

---

## 2. Text Generation
Text Generation adalah metode yang memungkinkan kita membuat teks berdasarkan input atau prompt tertentu. Menurut saya, ini adalah fitur yang sangat kreatif dan fleksibel, terutama dalam menghasilkan ide atau konten baru, seperti artikel, cerita, atau dialog chatbot. Text generation dapat sangat berguna dalam otomatisasi konten dan juga untuk pembuatan prototipe cepat bagi aplikasi yang membutuhkan interaksi berbasis teks.

---

## 3. Custom Text Generation with Specific Model and Parameters
Modifikasi Text Generation ini menggunakan model yang lebih ringan seperti DistilGPT2, dengan kemampuan untuk mengatur panjang dan jumlah respons. Pengaturan ini bermanfaat dalam skenario seperti asisten virtual atau pembuatan respons otomatis, di mana variasi dan efisiensi menjadi kunci dalam menjaga interaksi yang tidak monoton.

---

## 4. Fill-mask Pipeline
Fill-mask adalah jenis NLP yang digunakan untuk memprediksi kata yang hilang atau “masking”. Ini sangat berguna untuk tugas-tugas seperti koreksi otomatis, penyelesaian kalimat, atau bahkan dalam latihan bahasa. Fill-mask pipeline bermanfaat di industri pendidikan atau penerjemahan, di mana kita bisa melatih model melengkapi kata atau frasa yang hilang berdasarkan konteks.


## 5. Named Entity Recognition (NER) Pipeline
NER digunakan untuk mengidentifikasi entitas penting dalam teks seperti nama orang, lokasi, atau organisasi. Ini sangat berguna dalam pengelolaan data, terutama untuk sistem yang memerlukan ekstraksi informasi spesifik dari dokumen besar. Dengan NER, kita dapat mengotomatisasi pengumpulan data penting, misalnya, mengidentifikasi merek atau nama produk dalam ulasan konsumen.

---

## 6. Question-Answering Pipeline with Custom Context and Question
Question-answering (QA) memungkinkan kita untuk mengekstrak informasi penting dari teks, dengan menyajikan pertanyaan dan konteks. Menurut saya, QA sangat bermanfaat untuk aplikasi yang membutuhkan jawaban otomatis berdasarkan dokumen atau basis data besar. QA sangat berguna untuk meningkatkan efisiensi dalam pencarian informasi spesifik.

---

## 7. Sentiment Analysis Pipeline
Sentiment Analysis digunakan untuk memahami emosi atau sentimen dalam sebuah teks. Ini sangat penting dalam analisis umpan balik pelanggan, seperti pada ulasan produk atau komentar media sosial. Dengan mengkategorikan teks sebagai positif atau negatif, kita dapat memperoleh wawasan penting tentang perasaan pelanggan atau tren pasar secara keseluruhan.

---

## 8. Summarization Pipeline
Summarization Pipeline memungkinkan kita meringkas teks panjang menjadi lebih singkat dan informatif. Ini sangat berguna untuk merangkum artikel berita, laporan, atau dokumen bisnis yang panjang. Dengan teknologi ini, kita bisa menghemat waktu dalam membaca atau mengolah informasi panjang menjadi ringkasan yang lebih mudah dipahami.

---

## 9. Translation Pipeline (Indonesian to English)
Translation Pipeline digunakan untuk menerjemahkan teks dari satu bahasa ke bahasa lain, dalam contoh ini dari Bahasa Indonesia ke Bahasa Inggris. Ini sangat relevan untuk aplikasi yang membutuhkan keterjangkauan bahasa, terutama dalam konteks komunikasi global atau industri yang berhubungan dengan pengguna internasional.
